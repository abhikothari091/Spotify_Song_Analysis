```python
import pandas as pd
import numpy as np
df = pd.read_csv('tracks.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['explicit'].value_counts('1')*100
```

```python
df.info()
```

```python
df.set_index('release_date',inplace=True)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df.head()
```

```python
# now top 10 songs based on popularity
top_10 = df[['name','artists','popularity']].sort_values('popularity',ascending = False).head(10)
top_10
```

```python
# now we check the 10 worst songs
bottom_10=df[['name','artists','popularity']].sort_values('popularity',ascending = True).head(10)
bottom_10
```

```python
df['duration'] = df['duration_ms'].apply(lambda x: round(x/1000))
```

```python
df.head()
```

```python
df.drop('duration_ms',axis = 1,inplace=True)
```

```python
df.head()
df.drop('id',axis = 1,inplace=True)
```

```python
df.drop(['mode','key','explicit'],axis=1,inplace=True)
```

```python
# now we find correlation between fields
df_corr = df.corr(method ="pearson")
```

```python
df_corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
heatmap = sns.heatmap(df_corr,annot=True,cmap = "inferno")
heatmap.set_title("Heatmap correlation between different variables")
```

```python
# now to find relation between two correlated variables, we sample the data
sample_df = df.sample(int(len(df)*0.004))
#we sampled the 0.4 percent of the data
```

```python
sample_df.shape
```

```python
# loudness and energt
plt.figure(figsize=(10,6))
reg_plot1 = sns.regplot(sample_df,y = 'loudness',x = 'energy',color = 'blue')
reg_plot1.set_title('Loudness vs energy')
```

##we can conclude as loudness increases energy increases

```python
plt.figure(figsize=(10,6))
reg_plot2 = sns.regplot(sample_df,y = 'popularity',x = 'acousticness',color = 'yellow')
reg_plot2.set_title('Popularity vs accousticness')
```

#it is not a very prominent but still with an increase in accousticness, the popularity decreases

```python
# now to find year specific values, we split the date into year
df['dates'] = df.index.get_level_values('release_date')
df.dates = pd.to_datetime(df.dates)
years = df.dates.dt.year
```

```python
sns.displot(years,kind='hist')
```

#as we can see number of songs went on increasing and around 1990's it  saw a fall and gained again to rally further

```python
#now for duration over the years
duration = df.duration
fig,ax = plt.subplots(figsize = (16,6))
plt.figure(figsize=(10,6))
bar = sns.barplot(x = years, y=duration, ax = ax)
bar.set_title("Duration over the years")
```

##as we can see son duration over these years has been more or less constant

```python
# now if we check for a lineplot
fig,ax = plt.subplots(figsize = (16,6))
plt.figure(figsize=(10,6))
line = sns.lineplot(x = years, y=duration, ax = ax)
line.set_title("Duration over the years")
```
